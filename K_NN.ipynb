```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
x =[
    [170,65],
    [160,58],
    [160,80],
    [150,45],
    [155,48],
    [165,55]
]

y = ['athlete','athlete','athlete','non-athlete','non-athlete','non-athlete']
```

```python
#step 2: test oint
text_point = [[167,58]]
```

```python
#step 3: train K-NN model
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x,y)
```

```python
#step 4: predict the class
prediction = knn.predict(text_point)
print("prediction: ",prediction[0])
```

```python
#step 5: visualization
#separate classes for plotting
athletes = [x[i] for i in range(len(y)) if y[i]=='athlete']
non_athletes = [x[i] for i in range(len(y)) if y[i]=='non-athlete']
```

```python
#convert to x,y for plotting
athletes_x = [i[0] for i in athletes]
athletes_y = [i[1] for i in athletes]
non_athletes_x = [i[0] for i in non_athletes]
non_athletes_y = [i[1] for i in non_athletes]
```

```python
#plotting
plt.scatter(athletes_x,athletes_y,color='blue',label='athlete')
plt.scatter(non_athletes_x,non_athletes_y,color='red',label='non-athlete')
plt.scatter(text_point[0][0],text_point[0][1],color='green',label='text_point')
plt.xlabel('height')
plt.ylabel('weight')
plt.title('K-NN Classification')
plt.legend()
plt.grid(True)
plt.show()
```
